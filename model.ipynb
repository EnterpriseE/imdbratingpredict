```python
#跑gridsearch 


#评分的标准，r2,mse这样
from sklearn.model_selection import TimeSeriesSplit,GridSearchCV, cross_val_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import make_scorer
from sklearn.metrics import mean_squared_log_error
from sklearn.metrics import max_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import median_absolute_error




from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score
from sklearn.metrics import average_precision_score
from sklearn.metrics import balanced_accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import zero_one_loss

# from sklearn.metrics import d2_absolute_error_score



import pandas as pd

def runModels(model_params,X,y):
#     tscv = TimeSeriesSplit(n_splits=3)
#     grid_search = GridSearchCV(clf, param_grid, cv=tscv.split(X))
    df_new = pd.DataFrame()
    scoring1 = {'r2':make_scorer(r2_score,greater_is_better=True),
               'MSE':make_scorer(mean_squared_error),
               "MSLE":make_scorer(mean_squared_log_error),
               "ME":make_scorer(max_error),
               "MedianAE:":make_scorer(median_absolute_error),
               "Mean:":make_scorer(mean_absolute_error),
               "MedianAE:":make_scorer(max_error),
#                "AES":make_scorer(d2_absolute_error_score)
              }
    
    
    scoring2 = {
                'accuracy':make_scorer(accuracy_score,greater_is_better=True),
               'auc': make_scorer(roc_auc_score, needs_proba=True, multi_class="ovr"),
#                "average_precision_score":make_scorer(average_precision_score),
#                "balanced_accuracy_score":make_scorer(balanced_accuracy_score),
        
               "f1_micro":make_scorer(f1_score,average='micro'),
#                "zero_one_loss:":make_scorer(zero_one_loss),
#                "MedianAE:":make_scorer(max_error),
#                "AES":make_scorer(d2_absolute_error_score)
              }
#     scorers = ['roc_auc', 'f1', 'balanced_accuracy', 'accuracy']
    
    
    scores = []
    a_lot_bests_scores = pd.DataFrame()
    a_lot_scores = pd.DataFrame()
    for i in arange (6, 9.5, 0.5):
        y1 = trySection(y,i)
        
        
        
        #scoring 看用哪一个scores,scoring1是regression, scoring2是classification
        for model_name, mp in model_params.items():
    #         clf =  GridSearchCV(mp['model'], mp['params'], cv=5,refit='MSE', return_train_score=False,scoring=scoring1)
    
    
            #cv是平均几等分
            clf =  GridSearchCV(mp['model'], mp['params'], cv=5,refit='accuracy', return_train_score=False,scoring=scoring2)

    #         clf =  GridSearchCV(mp['model'], mp['params'], cv=5,refit='accuracy', return_train_score=False)

            clf.fit(X, y1)

        
            #拿到model的最好值，比如model最好的mse,最好的参数
            scores.append({
                "middle":i,
                'model': model_name,
                'best_score': clf.best_score_,
                'best_estimator':clf.best_estimator_,
                'best_params': clf.best_params_,

            })

            print(i)
            print("model:  "+model_name+'best_score:'+ str(clf.best_score_))
            print(clf.best_params_)

    #         c = cross_val_score(clf, X, y) 

            re = pd.DataFrame(clf.cv_results_)

            df_new = df_new.append(re,ignore_index=True)
            df_new["ind"] = i
    #         print(df_new)
            




        df_scores = pd.DataFrame(scores,columns=['middle','model','best_score','best_estimator','best_params'])
        a_lot_bests_scores.append(a_lot_bests_scores,ignore_index=True)
        a_lot_bests_scores["ind"] = i
    return [df_new,a_lot_bests_scores]
```
